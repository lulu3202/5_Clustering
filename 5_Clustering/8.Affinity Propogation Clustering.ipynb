```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AffinityPropagation.html

from sklearn.cluster import AffinityPropagation
import matplotlib.pyplot as plt

# Initialize the AffinityPropagation model
#affinity_prop = AffinityPropagation(random_state=42)
#affinity_prop = AffinityPropagation(affinity='euclidean')
affinity_prop = AffinityPropagation(max_iter=500)

# Fit the AffinityPropagation model to the data
affinity_prop.fit(X)

# Get the cluster labels for each data point
labels = affinity_prop.predict(X)

# Plot the data points and color them according to their cluster labels
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('Affinity Propagation Clustering')
plt.show()
```
