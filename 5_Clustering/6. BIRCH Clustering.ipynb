```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.Birch.html

from sklearn.cluster import Birch
import matplotlib.pyplot as plt

# Initialize BIRCH with a set number of clusters (you can leave it to None to determine automatically)
#birch_model = Birch(n_clusters=4)
#birch_model = Birch(threshold=0.5, n_clusters=4)
birch_model = Birch(threshold=0.5, n_clusters=5, branching_factor=50)

# Fit the BIRCH model to the data
birch_model.fit(X)

# Predict the cluster for each data point
label = birch_model.predict(X)
```

```python
supervised=pd.DataFrame(dataset)
```

```python

supervised['Cluster_group']=label
```

```python
supervised
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```
