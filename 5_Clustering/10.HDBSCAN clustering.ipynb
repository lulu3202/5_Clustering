```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
from sklearn.cluster import HDBSCAN
import matplotlib.pyplot as plt

# Initialize the HDBSCAN model with minimum cluster size
hdbscan_model = HDBSCAN(min_cluster_size=10)

# Fit the HDBSCAN model and predict cluster labels
labels = hdbscan_model.fit_predict(X)

# Plot the data points and color them according to their cluster labels
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('HDBSCAN Clustering')
plt.show()
```
