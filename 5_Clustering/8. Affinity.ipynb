```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.AffinityPropagation.html

from sklearn.cluster import AffinityPropagation
import matplotlib.pyplot as plt

# Initialize the AffinityPropagation model
#affinity_prop = AffinityPropagation(random_state=42)
#affinity_prop = AffinityPropagation(affinity='euclidean')
affinity_prop = AffinityPropagation(max_iter=500)

# Fit the AffinityPropagation model to the data
affinity_prop.fit(X)

# Get the cluster labels for each data point
label = affinity_prop.predict(X)

```

```python
supervised=pd.DataFrame(dataset)
```

```python

supervised['Cluster_group']=label
```

```python
supervised
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```
