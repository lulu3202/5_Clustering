```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
#https://drive.google.com/file/d/1ZLtNKvv8EvthVeWPYcgmLdJqfPEklkV3/view?usp=sharing

from sklearn.cluster import OPTICS
import matplotlib.pyplot as plt

# Initialize the OPTICS model
#optics_model = OPTICS(min_samples=15)
optics_model = OPTICS(min_samples=14,metric='manhattan')

# Fit the OPTICS model and predict cluster labels
label = optics_model.fit_predict(X)
```

```python
supervised=pd.DataFrame(dataset)
```

```python

supervised['Cluster_group']=label
```

```python
supervised
```

```python
import seaborn as sns
facet = sns.lmplot(data=supervised, x=supervised.columns[3], y=supervised.columns[4], hue=supervised.columns[5], 
                   fit_reg=False, legend=True, legend_out=True)
```
