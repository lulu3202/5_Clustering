```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
#https://scikit-learn.org/stable/modules/generated/sklearn.cluster.Birch.html

from sklearn.cluster import Birch
import matplotlib.pyplot as plt

# Initialize BIRCH with a set number of clusters (you can leave it to None to determine automatically)
#birch_model = Birch(n_clusters=4)
#birch_model = Birch(threshold=0.5, n_clusters=4)
birch_model = Birch(threshold=0.5, n_clusters=4, branching_factor=50)

# Fit the BIRCH model to the data
birch_model.fit(X)

# Predict the cluster for each data point
labels = birch_model.predict(X)

# Plot the data points and color them according to their cluster labels
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('BIRCH Clustering')
plt.show()
```
