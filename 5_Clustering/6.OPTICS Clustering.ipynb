```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3, 4]].values
```

```python
#https://drive.google.com/file/d/1ZLtNKvv8EvthVeWPYcgmLdJqfPEklkV3/view?usp=sharing

from sklearn.cluster import OPTICS
import matplotlib.pyplot as plt

# Initialize the OPTICS model
#optics_model = OPTICS(min_samples=5)
optics_model = OPTICS(min_samples=5,metric='manhattan')

# Fit the OPTICS model and predict cluster labels
labels = optics_model.fit_predict(X)

# Plot the data points and color them according to their cluster labels
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.title('OPTICS Clustering')
plt.show()
```
